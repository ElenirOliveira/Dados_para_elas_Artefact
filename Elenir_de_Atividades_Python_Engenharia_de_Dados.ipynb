##Atividade 1 - Média Salarial dos Departamentos
**Objetivo**: Desenvolver um script capaz de filtrar uma lista de funcionários por um departamento específico e, em seguida, calcular a média salarial apenas dos funcionários daquele departamento.

```python
# Estrutura de Dados de Entrada (Exemplo)
funcionarios = [
    {'nome': 'Ana', 'departamento': 'Vendas', 'salario': 3500.00},
    {'nome': 'Bruno', 'departamento': 'TI', 'salario': 6000.00},
    {'nome': 'Carla', 'departamento': 'Vendas', 'salario': 4000.00},
    {'nome': 'Daniel', 'departamento': 'RH', 'salario': 4500.00},
    {'nome': 'Elisa', 'departamento': 'TI', 'salario': 7500.00},
    {'nome': 'Felipe', 'departamento': 'Vendas', 'salario': 3800.00},
]


#Desenvolva seu código a partir daqui (dica: pense em como fazer seu código usando uma função)
```

```python
salarios = [f["salario"] for f in funcionarios if f["departamento"] == departamento_v]

print(salarios)  # Mostra os salários encontrados

if len(salarios) > 0:
    media = sum(salarios) / len(salarios)
    print(f"A média salarial do departamento {departamento_v} é {media:.2f}R$")
else:
    print("Nenhum funcionário encontrado nesse departamento.")
```

##Atividade 2 - ETL alunos aprovados
**Objetivo**: Desenvolver um script que leia um arquivo CSV com dados de alunos, adicione uma coluna de 'status' (Aprovado/Reprovado) com base nas regras abaixo e salve o resultado em um novo arquivo CSV.

Para o aluno ser **considerado Aprovado**: Nota >= 7.0  **e** Faltas <= 10

OBS: Usar a biblioteca **pandas**

```python
import pandas as pd
```

```python
import io

# Para simular o arquivo CSV sem precisar criá-lo fisicamente, vamos usar um texto de múltiplas linhas.
dados_csv = """nome,nota,faltas
Ana Silva,8.5,2
Bruno Costa,6.2,5
Carla Dias,9.0,12
Daniel Farias,4.5,3
Elisa Borges,7.0,7
Fernanda Lima,5.0,11
"""

# O io.StringIO permite que o pandas leia o texto como se fosse um arquivo.
arquivo_csv = pd.read_csv(io.StringIO(dados_csv))

#Desenvolva seu código a partir daqui (dica: pense em como fazer seu código usando uma função e nao esqueça de importar as bibliotecas)
```

```python
print(arquivo_csv)
```

```python
# Apply the condition to create the 'status' column
arquivo_csv['status'] = arquivo_csv.apply(lambda row: 'Aprovado' if row['nota'] >= 7.0 and row['faltas'] <= 10 else 'Reprovado', axis=1)

# Display the updated DataFrame
print(arquivo_csv)
```

```python
import numpy as np

# Apply the condition using np.where
arquivo_csv['status_np'] = np.where((arquivo_csv['nota'] >= 7.0) & (arquivo_csv['faltas'] <= 10), 'Aprovado', 'Reprovado')

# Display the updated DataFrame with the new column
print(arquivo_csv)
```

## Visualizar Média Salarial por Departamento

### Subtask:
Gerar um gráfico de barras mostrando a média salarial de cada departamento para responder a uma das perguntas.


```python
import matplotlib.pyplot as plt

salarios_por_departamento = {}
for funcionario in funcionarios:
    departamento = funcionario['departamento']
    salario = funcionario['salario']
    if departamento not in salarios_por_departamento:
        salarios_por_departamento[departamento] = []
    salarios_por_departamento[departamento].append(salario)

media_salarial_departamentos = {}
for departamento, salarios in salarios_por_departamento.items():
    if salarios:
        media_salarial_departamentos[departamento] = sum(salarios) / len(salarios)
    else:
        media_salarial_departamentos[departamento] = 0.0

print("Média salarial por departamento:", media_salarial_departamentos)
```

```python
departamentos = list(media_salarial_departamentos.keys())
medias_salariais = list(media_salarial_departamentos.values())

plt.figure(figsize=(10, 6))
plt.bar(departamentos, medias_salariais, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Departamento')
plt.ylabel('Média Salarial (R$)')
plt.title('Média Salarial por Departamento')
plt.grid(axis='y', linestyle='--')
plt.show()
```

## Visualizar Contagem de Alunos Aprovados/Reprovados

### Subtask:
Gerar um gráfico de barras mostrando a contagem de alunos aprovados e reprovados para responder a uma das perguntas.

```python
import matplotlib.pyplot as plt

status_counts = arquivo_csv['status'].value_counts()

print("Contagem de alunos por status:")
print(status_counts)
```

```python
status_labels = status_counts.index
status_values = status_counts.values

plt.figure(figsize=(7, 5))
plt.bar(status_labels, status_values, color=['lightcoral', 'skyblue'])
plt.xlabel('Status de Aprovação')
plt.ylabel('Número de Alunos')
plt.title('Contagem de Alunos Aprovados vs. Reprovados')
plt.grid(axis='y', linestyle='--')
plt.show()
```
